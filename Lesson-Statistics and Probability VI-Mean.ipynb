## Mean as a Balance Point

Above, we have seen that mean is not at the center of distribution's  0 - 13 range interval. 


In some cases, mean can be at the center of the range, but often it is not at the center. 

On the other hand, as we have seen above, we can correctly think of mean as lying at a point from where the sum of distances of points lying below that point and the sum of distances of points lying above it are equal. This is true for the distribution of any variable measured on an interval or ratio scale.




Mean can be described as the balance point of a lever. If the total distances above and below the mean were equivalent to the forces exerted by the weights on the rod of the lever, then there would be the same amount of force exerted on each side of the mean. This will make the rod stay in perfect equilibrium:


To verify, this peoperty of mean, below, we will:
Generate 5000 different distributions, measure the total distances above and below the mean, and check whether they are equal.
We will, for each of the 5000 iterations of a for loop:

- Set a seed using the `seed()` [function](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.seed.html) from `numpy.random`. For the first iteration, the seed number should be 0, for the second iteration it should be 1, for the third it should be 2, and so on.
- Generate randomly a distribution of integers using the `randint()` [function](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.randint.html) from `numpy.random`. Pass the right arguments to randint() such that each distribution will:
    - Have 10 values.
    - The values can range from 0 to 1000.
- Compute the mean of the distribution.
- Measure the total distance above and below the mean.
- Round off each distance to 1 decimal place using the round() function. This will prevent rounding errors at the 13th or 14th decimal place.
- Compare the two sums. If they are equal, then increment a variable named `equal_distances` with 1. We'll need to define `equal_distances` outside the loop with a value of 0.
- At the end `equal_distances` should have a value of 5000. This will confirm that for each of the 5000 distributions the total distance of the values above the mean is equal to the total distance of the values below the mean.

```python
import numpy as np
from numpy.random import randint, seed

equal_distances = 0

for i in range (0, 5000):
    np.random.seed(i)
    distribution = np.random.randint(low = 0, high = 1000, size = 10)
    mean = np.mean(distribution)
    
    above = []
    below = []
    for value in distribution:
        if value == mean:
            continue # continue with next iteration as distance is zero
        if value < mean:
            below.append(mean - value)
        else:
            above.append(value - mean)
            
    sum_above = round(sum(above), 1)
    sum_below = round(sum(below), 1)
    
    if (sum_above == sum_below):
        equal_distances += 1   
print(equal_distances)        
```

## Defining the Mean Algebraically

N values (N is the number of values in the population):

$$\begin{equation}
population \ mean = \frac{x_1 + x_2 + ... + x_N}{N}
\end{equation}$$

By convention, the mean of a population is denoted with the Greek letter   (pronounced "mew"). So we rewrite the formula above:

$$\begin{equation}
\mu = \frac{x_1 + x_2 + ... + x_N}{N}
\end{equation}$$

### Alternative Definition of Mean

Above, we defined population mean as:
$$\begin{equation}
\mu = \frac{x_1 + x_2 + ... + x_N}{N}
\end{equation}$$

$x_1 + x_2 + ... + x_N$ is often written as $\displaystyle\sum_{i = 1}^{N} x_i$

$$\begin{equation}
\mu = \frac{x_1 + x_2 + ... + x_N}{N} = \frac{\displaystyle\sum_{i = 1}^{N} x_i}{N}
\end{equation}$$

For a distribution:
D = [2, 4, 6]
the sum of the distribution D is:
$$\begin{equation}
\sum D = 2 + 4 + 6 = 12
\end{equation}$$

We can think of $\displaystyle\sum_{i = 1}^{N} x_i$ as a for loop.
where the iteration variable is i. This means that i  will take a different value for each iteration. 
i = 1 defines the starting value of the loop, which is 1. For every new iteration, the previous value of i is incremented by 1.

This is one way we could code in Python a similar logic:


Algebra is quite flexible, and following expression are equivalent:

$\frac{\displaystyle\sum_{i = 1}^{n} x_i}{n}$ = $\frac{1}{n}\  (\displaystyle\sum_{i = 1}^{n}x_i$

Below, we will:
Write a function that behaves just like $\frac{\displaystyle\sum_{i = 1}^{N} x_i}{N}$

- The function takes in an array of numbers and returns its mean value.
- Inside the function, write a for loop to iterate over the values in the array and sum them up. We can also use sum() to compute the sum of an array without using a for loop, but we will use a for loop to understand the logic of $\frac{\displaystyle\sum_{i = 1}^{N} x_i}{N}$
- Use the function to compute the mean of the three distributions given below.

```python
distribution_1 = [42, 24, 32, 11]
distribution_2 = [102, 32, 74, 15, 38, 45, 22]
distribution_3 = [3, 12, 7, 2, 15, 1, 21]

def get_mean(array):
    sum_array = 0
    for i in range(len(array)):
        sum_array += array[i]
    mean_array = sum_array / len(array)
    return mean_array
mean_1 = get_mean(distribution_1)
print(mean_1)
mean_2 = get_mean(distribution_2)
print(mean_2)
mean_3 = get_mean(distribution_3)
print(mean_3)
```

### Estimates from Low-Sized Samples

To prove this point, in the code below we will:

### Sample Mean as an Unbiased Estimator

If we took all the possible samples for a given sample size, we'd observe that the mean of the sample means will be equal to the population mean.

For:
X = [0, 3, 6]

mean = 3

and mean of means for all possible sample, if we take sample_size = 2 is also 3





$$\begin{equation}
\frac{1.5 + 3 + 1.5 + 4.5 + 3 + 4.5}{6} = \frac{18}{6} = 3 
\end{equation}$$

When a statistic is on average equal to the parameter it estimates, we call that statistic an **unbiased estimator**. In our case, the sample mean is an unbiased estimator of population mean.

This also holds true if we sample with replacement. When we do sampling with replacement, we sample one value, and then we put it back in the population, which means we can sample that value again. For instance, if we want a sample of size 2 from the population above, and we sample with replacement, this could happen:

- We extract one value randomly and get a 3.
- Because we sample with replacement, we put the value back in the population.
- We extract one more value and get a 3 again. We end up with this sample: [3, 3]

Below we can see the samples of size 2 we can get when we sample with replacement from the population above. We also show the mean for each sample:



The mean of the sample means amounts to 3 again and confirms that the sample mean is an unbiased estimator of population mean.


$$\begin{equation}
\frac{1.5 + 3 + 0 + 1.5 + 4.5 + 3 + 3 + 4.5 + 6}{9} = \frac{27}{9} = 3 
\end{equation}$$


Below, we Check whether the population mean of the population [3, 7, 2] is equal to the mean of all the sample means of size 2 that we can get if we do sampling without replacement.

- Compute the mean for each sample.
- Compute the mean of all the sample means. Compare it with the population mean using the == operator, and assign the result of the comparison to a variable named unbiased.

```python
population = [3, 7, 2]
samples = [[3, 7], [3, 2],
           [7, 2], [7, 3],
           [2, 3], [2, 7]
          ]
population_mean = sum(population) / len(population)
sample_means = []

for sample in samples:
    sample_mean = sum(sample)/ len(sample)
    sample_means.append(sample_mean)
    
mean_of_sample_means = sum(sample_means)/len(sample_means)    
unbiased = (population_mean == mean_of_sample_means)
unbiased
```
